# 공분산과 상관관계 실습

```python
# 패키지 설치

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# 데이터 불러오기
url = "https://raw.githubusercontent.com/c-karl/DA_DS_Book001/refs/heads/main/datasets/wine-quality.csv"
df = pd.read_csv(url)

# 데이터 샘플 확인
df.head()

## type 변수를 제외하고는 모두 숫자형 변수이므로 공분산과 상관계수를 분석하기에 적합
```

```python
# 산점도 행렬 시각화
sns.set(font_scale=1.1) ## 폰트 크기 설정
sns.set_style('ticks') ## 축 눈금 설정
sns.pairplot(df, diag_kind='kde') ## 산점도 행렬 시각화
                                  ## diag_kind='kde' 옵션은 동일한 변수의 산점도를 분포로 시각화
plt.show()
```

```python
# 공분산 확인
df.cov(numeric_only=True)

## 공분산은 직관적으로 상관성의 높고 낮음을 파악하기 힘들다.
```

```python
# 피어슨 상관계수 확인
corr = df.corr(method='pearson', numeric_only=True)
corr
```

```python
# 히트맵 시각화
sns.heatmap(corr, cmap='viridis')

## 정확한 수치가 나타나지 않음
```

```python
# 히트맵 시각화 2
sns.clustermap(corr, annot=True, cmap='RdYlBu_r', vmin=-1, vmax=1)
```

```python
# 히트맵 시각화 3
# 중복 제거

## 매트릭스 우측 상단을 모두 True인 1로, 하단을 False인 0로 변환
np.triu(np.ones_like(corr))

## True/False mask 배열로 변환.
mask = np.triu(np.ones_like(corr, dtype=bool))

## 히트맵 그래프
fig, ax = plt.subplots(figsize=(15, 10))
sns.heatmap(corr, mask=mask, annot=True, cmap='RdYlBu_r', vmin=-1, vmax=1, cbar=True)
ax.set_title('Wine Quality Correlation', pad=15)
```
