# 시간 시각화 실습

```python
# 패키지 설치

import matplotlib.pyplot as plt
import pandas as pd
import datetime
```

```python
# 데이터 불러오기
url = "https://raw.githubusercontent.com/c-karl/DA_DS_Book001/refs/heads/main/datasets/superstore.csv"
df = pd.read_csv(url)

# 데이터 샘플 확인
df.head()
```

```python
# 선 그래프 시각화를 위한 데이터 가공

## date 컬럼 날짜형식 변환
from datetime import datetime  # datetime 모듈에서 datetime 클래스 가져오기

def parse_date_by_range(date_str):
    parts = date_str.split('/')
    if int(parts[0]) > 12:  # 첫 번째 값이 12를 초과하면 %d/%m/%Y로 가정
        return datetime.strptime(date_str, "%d/%m/%Y")
    else:  # 그렇지 않으면 %m/%d/%Y로 가정
        return datetime.strptime(date_str, "%m/%d/%Y")

df['Order_Date2'] = df['Order Date'].apply(parse_date_by_range)
## 날짜 오름차순 정렬
df = df.sort_values(by='Order_Date2')
## 연도 컬럼 생성
df['Year'] = df['Order_Date2'].dt.year

# 선 그래프 용 데이터셋 생성

## 2018년 데이터만 필터링
df_line=df[df.Year == 2018]
## 2018년 일 별 매출액 가공
df_line = df_line.groupby('Order_Date2')['Sales'].sum().reset_index()

df_line.head()
```

```python
# 30일 이동평균 생성
df_line['Month'] = df_line['Sales'].rolling(window=30).mean()

# 선 그래프 시각화
ax = df_line.plot(x='Order_Date2', y='Sales',linewidth = "0.5")
df_line.plot(x='Order_Date2', y='Month', color='#FF7F50', linewidth = "1", ax=ax)

## 5월부터 매출이 감소했다가 8월부터 상승하는 것을 확인
## 편차가 큰 변수를 확인하고자 하는 경우에는 이동평균선을 활용하는 것이 도움이 됨.
```

```python
# 막대그래프 시각화를 위한 데이터 가공

## 연도 별 판매량 데이터 가공
df_bar_1 = df.groupby('Year')['Sales'].sum().reset_index()
df_bar_1.head()
```

```python
# 연도 별 매출액 막대 그래프 시각화
ax = df_bar_1.plot.bar(x='Year', y='Sales', rot=90, figsize=(10,5))
```

```python
#연도별, 고객 세그먼트 별 매출액 데이터 가공
df_bar_2 = df.groupby(['Year', 'Segment'])['Sales'].sum().reset_index()

# 고객 세그먼트를 컬럼으로 피벗
df_bar_2_pv = df_bar_2.pivot(index='Year', columns='Segment', values='Sales').reset_index()
df_bar_2_pv.head()
```

```python
# 연도 별 고객 세그먼트 별 매출액 누적 막대 그래프 시각화
# stacked=True 옵션을 통해 누적 막대 그래프로 시각화
df_bar_2_pv.plot.bar(x='Year', stacked=True, figsize=(10,7))
```
